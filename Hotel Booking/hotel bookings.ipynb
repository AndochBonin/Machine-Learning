```python
from matplotlib import pyplot as plot
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
document = pd.read_csv(r'C:\Users\bonin\Downloads\booking_of_hotel.csv')
```

```python
df = pd.DataFrame(document)
```

```python
df.head()
```

```python
number_of_hotel_bookings = len(df.index)
print(number_of_hotel_bookings)
```

```python
number_of_cancellations = len(df[df['is_canceled'] != 0])
print(number_of_cancellations)
```

```python
number_of_noncancellations = len(df[df['is_canceled'] == 0])
print(number_of_noncancellations)
```

```python
cancel_array = [number_of_cancellations, number_of_noncancellations]
```

```python
plot.bar(["Cancelled", "Not Cancelled"], cancel_array)
plot.title("Cancellations Bar Chart")
```

```python
most_popular_month_of_arrival = df["arrival_date_month"].mode()
print(most_popular_month_of_arrival)
```
